```python
# code for creating choropleth map of USA states
# import plotly library
import plotly

# import plotly.express module
# this module is used to create entire figures at once
import plotly.express as px

# create figure
fig = px.choropleth(pd_US,locations = 'State',
                    template = 'plotly_dark',
              color_continuous_scale= 'viridis',
                    basemap_visible = False,
                    height=600,width=800,
                    locationmode="USA-states",color= 'Confirmed',  scope="usa",title='COVID Positive Cases in USA',hover_name = 'State',
                    labels={'Confirmed':'Confirmed Covid Cases','State':'State Code'})

fig.show()
```
